### 二值化

```python
def square_error(d1, d2):
    '返回二类聚类的平方损失判据，归一到0-1'
    n1 = sum([i[-1] for i in d1])
    n2 = sum([i[-1] for i in d2])
    s1 = sum([k*v for k,v in d1])
    s2 = sum([k*v for k,v in d2])
    m1 = s1 / n1
    m2 = s2 / n2
    m0 = (s1 + s2) / (n1 + n2)
    e1 = sum([(k-m1)**2*v for k,v in d1])
    e2 = sum([(k-m2)**2*v for k,v in d2])
    e0 = sum([(k-m0)**2*v for k,v in d1]) + sum([(k-m0)**2*v for k,v in d2])
    return (e1 + e2) / e0
```

```python
from collections import Counter
```

```python
def devide(pic):
    '返回一维二类聚类的分界值，小于一类，大于等于一类'
    p = pic.flatten()
    c = sorted(Counter(p).items())
    es = []
    for i in range(1, len(c)-1):
        es.append(square_error(c[:i], c[i:]))
    ind = es.index(min(es))
    return c[ind+1][0]
```

```python
def binarize(p, th='auto'): 
    if th == 'auto':
        th = devide(p)
        print(f'二值化阈值取：{th}')
    return np.where(p<th, 0, 1)
```

```python
def backcolor_of(pic):
    '输入0-1二值图片，返回0或1代表背景色，判据是内聚度低为背景'
    m = [np.array([0,0], np.float64), np.array([0,0], np.float64)]
    s = [np.array([0,0], np.float64), np.array([0,0], np.float64)]
    n = [0, 0]
    for r in range(pic.shape[0]):
        for c in range(pic.shape[1]):
            i = pic[r, c]
            m[i] += (r, c)
            s[i] += (r**2, c**2)
            n[i] += 1
    e = [sum(s)/n-sum((m/n)**2) for m,s,n in zip(m,s,n)]
    bc = 1 if e[0] < e[1] else 0
    print(f'背景色为{bc}，两种颜色的平方误差分别为：{e}')
    return bc
```

```python
def strip(p, backgroundColor=1):
    isbak = (p == backgroundColor)
    p = p[~isbak.all(axis=1)]
    p = p[:, ~isbak.all(axis=0)]
    return p
```

```python
def preprocess(p, kn=7, th='auto', set_bakc=1):
    bp = binarize(gray(p), th)
    bakc = backcolor_of(bp)
    if bakc != 1:
        p = 1 - p
    fp = medfilt(gray(p), kn)
    if 1 != set_bakc:
        fp = 1 - fp
    bp = binarize(fp, th)
    return strip(bp, set_bakc)
```

## 边缘检测

```python
kn = [[1,1,1],[1,0,1],[1,1,1]]
def bound(p):
    'the bound pixels is 1'
    mp = signal.convolve2d(p, kn, mode='same', fillvalue=1)
    mp = np.where((mp>=1)&(p==0), 1, 0)
    return mp
```

### 函数

```python
def Chinese_Postman(G, min_accepted_len = 5):
    odd = {v:d for v,d in G.degree if d%2==1}#奇数度顶点
    nds = list(odd.keys())
    paths = []
    for i in range(0, len(nds), 2):
        paths.append(nx.shortest_path(G, nds[i], nds[i+1]))#两两配对画最短路
    pathG = nx.MultiGraph()
    for p in paths:
        pathG.add_path(p)#所有path组成一个图
    pegs = [e[:2] for e in pathG.edges]
    pegs = [e if e[0]<=e[1] else e[::-1] for e in pegs]
    c = Counter(pegs)
    pegs = [k for k,v in c.items() if v%2==1] #最多出现一次，出现偶数次对消  
    pathG = nx.Graph()
    pathG.add_edges_from(pegs)
    
    cycles = nx.cycle_basis(G)
    for cycle in cycles:
        cyG = nx.Graph()
        nx.add_cycle(cyG, cycle)
        cyE = {e if e[0]<=e[1] else e[::-1] for e in cyG.edges}
        in_cycle = cyE & pathG.edges
        if len(in_cycle) > len(cyE)/2: #圈中有超出一半的边要添加则取反
            pathG.remove_edges_from(in_cycle)
            pathG.add_edges_from((cyE-in_cycle))
            
    mG = nx.MultiGraph(G)#给G添重边
    mG.add_edges_from(pathG.edges)
    eulerian_circuit = [p[0] for p in nx.eulerian_circuit(mG)]
    return eulerian_circuit
```

```python
def find_eulerian_circuit(p, min_accepted_len=5):
    idim = gh.to_id_image(p)
    pairs = gh.neiPairsOf(idim)
    G = nx.Graph()
    G.add_edges_from(pairs)
    largest_cc = max(nx.connected_components(G), key=len)
    G = G.subgraph(largest_cc) #只取图中最大的连续边缘
    euler_cir = Chinese_Postman(G, min_accepted_len)
    return euler_cir
```

```python
def to_xy(ids, shape):
    coords = gh.toCoords(ids, shape)
    y, x = coords.T
    y = p.shape[0] - y
    return x, y
```

```python
from matplotlib.patches import Circle
from matplotlib.collections import PatchCollection
```

```python
def add_circles(ax, cs):
    p = PatchCollection(cs, edgecolors='0.2', facecolors='none', linewidths=1)
    ax.add_collection(p)
```

```python
patches=[]
for i in range(1,4):
    circle = Circle((0, i), i)
    patches.append(circle)
```

```python
fig, ax = plt.subplots()    
add_circles(ax, patches)
ax.axis('scaled')
```

```python
w0 = 2 * np.pi / tl
fr2 = freq[mask]
sp2 = sp[mask]/tl
```

```python
si = np.argsort(abs(sp2)[1:])[::-1]
si = [0, *(si+1)]
fr2 = fr2[si]
sp2 = sp2[si]
maxfre = int(max(abs(fr2)))
n_frame = 4 * maxfre
```

```python
i = np.c_[:n_frame]
```

```python
frame = sp2 * np.exp(1j*i*fr2/n_frame*np.pi*2)
```

```python
frame.shape
```

```python
rr = abs(frame)[:,1:]
oo = frame.cumsum(axis=-1)
```

```python
def get_cirs(rr, oo):
    circles=[]
    for rf, of in zip(rr, oo):
        cirs = []
        for ri, oi in zip(rf, of):
    #         print(of.real, of.imag)
            circle = Circle((oi.real, oi.imag), ri)
            cirs.append(circle)
        circles.append(cirs)
    return circles
```

```python
circles = get_cirs(rr, oo)
```

```python
xm, xM = iff.real.min(), iff.real.max()
ym, yM = iff.imag.min(), iff.imag.max()
xs, ys = xM - xm, yM - ym
r = 0.15
xl, xr = xm-r*xs, xM+r*xs
yb, yt = ym-r*ys, yM+r*ys
figsize=(10, 10*(yt-yb)/(xr-xl))
```

```python
def plot_frame(ax, i, circles, oo, trace, n_frame):
    cs = circles[i]
    os = oo[i]   
    add_circles(ax, cs)
    ax.plot(os.real, os.imag, 'k')
    ax.plot(os.real, os.imag, 'k.')
    ax.plot(os.real[-1], os.imag[-1], 'o', c='#e74c3c')
    ax.plot(os.real[0], os.imag[0], 'ko')
    ti = int(i/n_frame*len(trace))
    plot(trace.real[:ti], trace.imag[:ti], lw=5, alpha=1, c='#e74c3c')
    ax.axis('equal')
    ax.axis('off')
    ax.set_xlim(xl, xr)
    ax.set_ylim(yb, yt)
    return ax
```

```python
fig, ax = plt.subplots(figsize=figsize)
plot_frame(ax, 72, circles, oo, iff, n_frame)
```

```python
bufs = []
circles = get_cirs(rr, oo)
fig, ax = plt.subplots(figsize=figsize)
for i in range(n_frame):
    buf = BytesIO()
    fig, ax = plt.subplots(figsize=figsize)
    plot_frame(ax, i, circles, oo, iff, n_frame)
    fig.savefig(buf)
    bufs.append(buf)
    plt.close(fig)
```
